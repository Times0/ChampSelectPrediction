```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
matches_df = pd.read_csv('2024_LoL_esports_match_data_from_OraclesElixir.csv')
grouped = matches_df.groupby('gameid')
```

```python
grouped.head()
```

```python
transformed_data = []
for name, group in grouped:
    # Get Blue and Red team data
    Blue_team = group[group["side"] == "Blue"]
    Red_team = group[group["side"] == "Red"]

    # Create a dictionary to store the transformed data for this game
    game_data = {
        "teamBlue": Blue_team.iloc[0]["teamname"],
        "teamRed": Red_team.iloc[0]["teamname"],
        "patch": group.iloc[0]["patch"],
        "winner": "Blue" if Blue_team.iloc[0]["result"] == 1 else "Red",
    }

    # Extract champion data for each team
    for i in range(1, 6):
        champ1 = Blue_team[Blue_team["participantid"] == i]["champion"].values[0]
        champ2 = Red_team[Red_team["participantid"] == i + 5]["champion"].values[0]
        game_data[f"Blue_champion_{i}"] = champ1
        game_data[f"Red_champion_{i}"] = champ2

        
            

    # Append the game data to the transformed data list
    transformed_data.append(game_data)


transformed_df = pd.DataFrame(transformed_data)

transformed_df.head()
```

```python
transformed_df.to_csv("transformed_data.csv", index=False)
```
